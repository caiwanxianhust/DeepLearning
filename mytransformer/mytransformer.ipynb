```python
import tensorflow as tf
from tensorflow import keras
from transformer import Transformer
```

```python
inp_vocab_size=5000
tar_vocab_size=5000
em_dim=128
maximum_position=80
num_layers=3
heads=8
rate=0.1
training=True
```

```python
transformer = Transformer(inp_vocab_size, 
                          tar_vocab_size, 
                          em_dim, 
                          maximum_position, 
                          num_layers, 
                          heads, 
                          rate,
                          training)
```

```python
transformer.transformer.summary()
```

```python
keras.utils.plot_model(transformer.transformer, "../model_img/mytransformer.png", 
                       show_layer_names=True, show_shapes="True")
```

```python
keras.utils.plot_model(transformer.decoder_model, "../model_img/mytransformer_decoder.png", 
                       show_layer_names=True, show_shapes="True")
```
